## Exercício de senha com um timer em Python.

```
import time
tempo_inicial = time.time()

senha = "54321"
leitura =" "

def acesso (leitura, senha):
  if leitura == senha:
    print('Acesso liberado ')
  else:
    print('Senha incorreta. Tente novamente')

while (leitura != senha):
  print(30 * "-")
  leitura = input("Digite a senha: ")
  tempoExec = time.time() - tempo_inicial
  print("Tempo de execução: {} segundos".format(tempoExec))
  if tempoExec <= 5:
    print(acesso(leitura, senha))
  else:
    print('Tempo excedido')
    break

# A linha time.time() retorna o tempo em segundos desde 1 de janeiro de 1970, 
# e a linha time.time() - t1 representa a variação de tempo, ou seja, quanto tempo o programa demorou para rodar.
```
