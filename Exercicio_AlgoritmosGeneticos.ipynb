```
import random
import datetime
```

```
#POPULAÇÃO
def _geracao_pais(comprimento, geneAjuste):
    genes = ''
    for i in range(0, comprimento, 1):
        genes = genes + geneAjuste[random.randint(0, len(geneAjuste) - 1)]
    return genes
```

```
#MUTAÇÃO
def _mutacao(pais, geneAjuste):
    indice = random.randint(0, len(pais) - 1)
    novoGene = geneAjuste[random.randint(0, len(geneAjuste) - 1)]
    while novoGene == pais[indice]:
        novoGene = geneAjuste[random.randint(0, len(geneAjuste) - 1)]
    proleGenes = ''
    for i in range(0, len(pais), 1):
        if i == indice:
          proleGenes = proleGenes + novoGene
        else: proleGenes = proleGenes + pais[i]
    return proleGenes
```

```
#CRUZAMENTO
def pegar_melhor(pegar_aptidao, alvoLen, otimaAptidao, geneAjuste, display):
    random.seed()
    melhoresPais = _geracao_pais(alvoLen, geneAjuste)
    melhorAptidao = pegar_aptidao(melhoresPais)
    display(melhoresPais)
    if melhorAptidao >= otimaAptidao:
        return melhoresPais
        
#MOTORGENÉTICO
    ct = 0
    while True:
        prole = _mutacao(melhoresPais, geneAjuste)
        proleAptidao = pegar_aptidao(prole)
        ct += 1
        if melhorAptidao < proleAptidao:
            if proleAptidao >= otimaAptidao:
                print('Iterações:', ct)
                return prole
            melhorAptidao = proleAptidao
            melhoresPais = prole
```

```
#MONITORAÇÃO
def display(genes, alvo, inicioTempo):
    tempoDif = datetime.datetime.now() - inicioTempo
    aptidao = pegar_aptidao(genes, alvo)
    print("{}\t{}\t{}".format(genes, aptidao, tempoDif))
```

```
#APTIDÃO
def pegar_aptidao(genes, alvo):
    tot = 0
    for i in range(len(alvo)):
        if alvo[i] == genes[i]:
            tot = tot + 1
    return tot
```

```
#CÓDIGO DA ROTINA QUE CHAMA O ALGORITMO GENÉTICO
def adivinhar_senha(alvo):
    geneajuste = '0123456789'
    inicioTempo = datetime.datetime.now()

    def fnPegarAptidao(genes):
        return pegar_aptidao(genes, alvo)

    def fnDisplay(genes):
        display(genes, alvo, inicioTempo)

    otimaAptidao = len(alvo)
    genes = pegar_melhor(fnPegarAptidao, len(alvo), otimaAptidao,geneajuste, fnDisplay)

    tempoFinal = datetime.datetime.now() - inicioTempo
    print('Tempo:', tempoFinal)
    return genes
```
