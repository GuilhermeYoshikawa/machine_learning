## 1) Escrever um programa para determinar a quantidade de cavalos necessários para levantar uma massa de m quilogramas a uma altura de h metros em t segundos, através da formula:


---


cavalos = m h/t 
          ____

          745,6999

```
m = float(input("Digite o valor da massa em quilogramas: "))
h = float(input("Digite o valor da altura em metros: "))
t = float(input("Digite o tempo em segundos: "))

cavalos = ((m*h)/t) / 745.6999

print("Quantidade de cavalos necessários: ", cavalos)
```

## 2) Fazer o diagrama de blocos para um programa que determine se uma pessoa é adulta, a partir do ano de nascimento e do ano atual. Exibir a frase:

---

"Voce e" situação " e " acao " ENTRAR."

Onde se idade for < 18:

> situacao é "menor" e

> acao é "nao pode"

Caso contrário:

> situacao é "maior" e

> acao é "pode"


```
nasc = int(input("Digite o ano de nascimento: "))
atual = 2020
situcao = " "

idade = atual - nasc 
print("Idade: ", idade, " anos")

if idade < 18:
  situacao = "menor"
  acao = "nao pode"
else:
  situacao = "maior"
  acao = "pode"

print("Voce e", situacao, " e ", acao, " ENTRAR.")
```

## 3) Ler a massa corporal e a altura de 20 pessoas. Para cada uma exibir seu IMC (Índice de massa corporal) e a sua classificação de acordo com a tabela abaixo:

---

IMC = massa
      _____

      altura**2

      

```
pessoas = 0

while pessoas < 20:
  pessoas = pessoas + 1
  print("Pessoa nº ", str(pessoas))
  massa = float(input("Digite sua massa corporal: "))
  altura = float(input("Digite sua altura: "))

  imc = massa / (altura ** 2)
  print("IMC: ", imc)

  if imc < 18.5:
    print("Abaixo do peso.")
  elif 18.5 <= imc < 24.9:
    print("Peso ideal (parabéns).")
  elif 25.0 <= imc < 29.9:
    print("Levemente acima do peso.")
  elif 30.0 <= imc < 34.9:
    print("Obesidade grau I.")
  elif 35.0 <= imc < 39.9:
    print("Obesidade grau II.")
  elif imc > 40:
    print("Obesidade grau III (mórbida).")
```

## 4) Ler o nome e a altura de 10 alunos e exibir o nome e a altura do mais baixo.

```
alunos = []
turma = []

for cont in range(1, 11):
  print("Aluno ", str(cont))
  alunos.append(float(input("Digite sua altura: ")))
  alunos.append(str(input("Digite seu nome: ")))
  turma.append(alunos[:])
  alunos.clear()

print("Menor aluno: {}".format(min(turma)))
```

## 5) Calcular a quantidade de numerário (cédulas e moedas) para pagar os funcionários de uma empresa em dinheiro, ignorando os centavos. Serão coletados os seguintes dados de cada funcionário:

### a) Número de identificação (0 indica que não há mais funcionários).
### b) Quantidade de horas trabalhadas.
### c) Salário hora.

## Determinar e exibir:

### 1. O valor total pago a todos os funcionários
### 2. A quantidade de numerário necessário a esse pagamento, ou seja, a quantidade de moedas de um real, notas de dois reais, cinco reais, e etc.

```
horasTrabalhadas = []
salarioHora = []
cont = 0
ident = " "
zero = "0"

while(ident != zero):
  print(30 * "-")
  cont = cont + 1
  print("Funcionário nº ", str(cont))
  ident = input("Número de identificação: ")
  if ident == zero:
    print(30 * "-")
    print("Não há mais funcionários.")
    break
  else:
    horasTrabalhadas.append(int(input("Quantidade de horas trabalhadas: ")))
    salarioHora.append(float(input("Salário hora: R$ ")))

print(30 * "-")
salario = (sum(horasTrabalhadas) * sum(salarioHora))
print("O valor total pago a todos os funcionários: R$ ", salario)

cem = int(salario / 100)
salario = salario - (cem*100)
      
cinquenta = int(salario/50)
salario = salario - (cinquenta*50)

vinte = int(salario/20)
salario = salario - (vinte*20)

dez = int(salario/10)
salario = salario - (dez*10)

cinco = int(salario/5)
salario = salario - (cinco*5)
  
um = salario

print('Cédulas de R$100,00 = ', cem)
print('Cédulas de R$ 50,00 = ', cinquenta)
print('Cédulas de R$ 20,00 = ', vinte)
print('Cédulas de R$ 10,00 = ', dez)
print('Cédulas de R$  5,00 = ', cinco)
print('Moedas de R$  1,00 = ', um)
```
