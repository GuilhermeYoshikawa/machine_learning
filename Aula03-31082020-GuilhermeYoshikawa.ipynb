````
d = {'gato': 'cat', 'cachorro': 'dog'}
# Cria um novo dicionário com alguns dados

print(d['gato'])
# Identifica significado de uma entrada e mostra "cat"
print('gato' in d)
# Verifica se uma chave está e mostra "True"
print(d)
d['peixe'] = 'fish'
# Inclui um element em um dicionário
print(d['peixe'])
# Mostra "fish"
print(d['macaco'])
# KeyError: 'macaco' porque não é uma chave do dicionário d
print(d.get('macaco', 'N/A'))
# Busca um elemento com valor default para falha, mostra "N/A"
print(d.get('peixe', 'N/A'))
# Busca um elemento com valor default para falha, mostra "fish"
del d['peixe']
# Remove um element do dicionário
print(d.get('peixe', 'N/A'))
# Busca um elemento com valor default para falha, mostra "N/A"
````

Primero é criado um novo dicionário.

print(d['gato']), mostra a tradução do conteúdo 'gato', retornando o 'cat'.

print('gato' in d), verifica 'gato' está em d, caso esteja, retorna "True", caso contrário, retorna "False".

d['peixe'] = 'fish', inclui um elemento no dicionário.

print(d['macaco']), mostra uma KeyError: 'macaco', porque não é uma chave do dicionário d.

E o del d['peixe'], remove o elemento do dicionário d.


````
animais = {'gato', 'cachorro'}
print('gato' in animais)
# Verifiva se um elemento está no set e mostra "True"
print('peixe' in animais)
# Mostra "False"
animais.add('peixe')
# Inclui um elemento no set
print('peixe' in animais)
# Mostra "True"
print(len(animais))
# Identifica o número de elementos do set e mostra "3"
animais.add('gato')
# Incluir un elemento que já existem não faz nada
print(len(animais))
# Mostra "3"
animais.remove('gato')
# Remove um elemento do set
print(len(animais))
# Mostra "2"
````

Primeiro cria um dicionário chamado, adicionando conteúdo dentro dele.

Com um "print('gato' in animais)", procura dentro do dicionário "animais" a palavra "gato". Caso seja encontrado ele retorna "True", caso contrário, retorna "False".

Com o exemplo "animais.add('peixe')", será includo o elemento 'peixe'.

O exemplo "print(len(animais))", mostra o tamanho do dicionário.

E o "animais.remove('gato')", excluí o elemento 'gato' do dicionário.

````
tp1 = (1,2,3,4,5,6)
# Cria uma tupla
print(tp1)
# Mostra a tupla (1,2,3,4,5,6)
print(tp1[4])
# Mostra o "5"
````

Primeiro é criado um dicionário dentro da variável tp1.

Depois com um "print(tp1)", é mostrado o dicionário completo.

E com um "print(tp1[4])", mostra onde o número que está na posição 5, com a contagem começando sempre do 0.

````
d = {(0, 1):0, (1, 2): 1, (2, 3): 2, (3, 4): 3, (4, 5): 4, 
     (5, 6): 5, (6, 7): 6, (7, 8): 7, (8, 9): 8, (9, 10): 9}
# Cria dicionários com tuplas como chaves
print(d)
# Mostra todas as tuplas do dicionário
tp2 = (5, 6)
# Cria uma tupla
print(d[tp2])
# Mostra "5"
print(d[(1, 2)])
# Mostra "1"
````

Cria uma tupla e sua respectiva chave dentro da variavel "d". 

No "print(d)" ele mostra o dicionário completo criado.

No "print(d[tp2])" mostra onde a tupla do tp2 (que é (5, 6)) está localizada dentro do dicionário d.

````
d = {(x, x + 1): x for x in range(10)}
# Cria dicionários com tuplas como chaves
print(d)
# Mostra todas as tuplas do dicionário
# {(0, 1): 0, (1, 2): 1, (2, 3): 2, (3, 4): 3, (4, 5): 4,
# (5, 6): 5, (6, 7): 6, (7, 8): 7, (8, 9): 8, (9, 10): 9}
tp2 = (5, 6)
# Cria uma tupla
print(type(tp2))
# Mostra "<class 'tuple'>"
print(d[tp2])
# Mostra "5"
print(d[(1, 2)])
# Mostra "1"
````

Na primeira parte, a iteração vai ficar 0 + 1, com o x que vai ser incrementado até tamanho 10, preenchendo automaticamente uma tupla sem precisar ficar preenchendo manualmente uma por uma.

O "print(type(tp2))" irá mostrar um "<class 'tuple'>".

"print(d[tp2])" mostra a chave que é encontrado a tupla tp2 (que é (5, 6)), dentro do dicionário d, retornando a chave 5.

O "print(d([1, 2]))" mostra a chave em que é encontrado a tupla (1, 2), dentro do dicionário d.

````
def leitura():
  nome = input("Digite seu nome: ")
  sobrenome = input("Digite seu sobrenome: ")
  nomecompleto = nome + " " + sobrenome
  print("Seu nome completo eh: ", nomecompleto)

leitura()
````

Neste caso é apresentado a função.

Primeiro é criado uma função, chamada leitura.

Dando um espaço para identação, mostra que todo conteúdo que estiver com um espaço, está dentro da função criada.

Depois foram adicionados 2 inputs, dentro dos atributos "nome" e "sobrenome", pedindo para inseri-los dentro do input.

Por último é listado o nome completo, chamando o atributo "nomecompleto", que puxa o nome e o sobrenome digitado.

Com o "leitura()" fora da identação, irá dar início a função criada.

````
def potencia(a,b):
  c = a ** b
  return c

x = int(input("Digite um valor: "))
y = int(input("Digite o valor de potência: "))
print(potencia(x,y))
````

Neste caso, foi criado a função "potencia" que recebe os valores "a" e "b". E retorna o "c", que é a representação do calculo de potência. Onde o "**" é o sinal de elevado.

Depois, fora da identação, é adicionado os valores através do int(input()), onde o int significa que o input vai receber valores númericos do tipo inteiro.

Após digitar os valores de a e b, será exibido o valor do calculo de potência, chamando a função potencia com os valores x e y.

````
def soma(a,b):
  c = a + b
  return c

x = float(input("Digite um valor: "))
y = float(input("Digite um valor: "))
print(soma(x,y))
````

Esse caso é bem parecido com o anterior, porém é trabalhado com um tipo de dado diferente.

Neste caso, o input recebe um elemento do tipo float, que são os números quebrados.

Digitando os valores de a e b, irá ser executado a função soma.
